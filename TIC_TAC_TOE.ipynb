```python
from IPython.display import clear_output
def display(gamelist):
    
    clear_output()
    print(' {} | {} | {} '.format(gamelist[1],gamelist[2],gamelist[3]))
    print('-----------')
    print(' {} | {} | {} '.format(gamelist[4],gamelist[5],gamelist[6]))
    print('-----------')
    print(' {} | {} | {} '.format(gamelist[7],gamelist[8],gamelist[9]))
```

```python
def position(gamelist,choice,player):
    pos = 0
    while(int(pos) not in range(1,10) or gamelist[int(pos)] == 'X' or gamelist[int(pos)] == 'O'):
        pos = input('Enter the position to add by player{}? '.format(player))
        if pos.isdigit() == False :
                pos = 0
                print('wrong form of input.')
             
    return int(pos)
```

```python
def game_over(game_list,choice):
    for k in [1,4,7]:
        if (game_list[k] == game_list[k+1] == game_list[k+2]== choice ):
                return False
     
    for k in range(0,4):
        if(game_list[k] == game_list[k+3] == game_list[k+6] == choice):
                return False
 
    if(game_list[1] == game_list[5] == game_list[9] == choice or game_list[3] == game_list[5] == game_list[7] == choice):
 
            return False
 
    return True   
```

```python
def replay():
    choice = input('Play again?(Y or N)').upper()
    return choice == 'Y'
```

```python
print('Welcome to TIC TAC TOE')
play = True
while(play):
    game_on = True
    game_list = [' ',' ',' ',' ',' ',' ',' ',' ',' ',' ']

    j = 0
# j is used to know when the game list is filled up
    choice_1 = '1'
# choice_1 is choice of player 1 and choice_2 is choice of player 2
# assign markers
    while choice_1 not in ['X','O']: 
        choice_1 = input("What is the shape for player 1 'X' or 'O'?").upper()
    if choice_1 == 'X':
        choice_2 = 'O'
    if choice_1 == 'O':
        choice_2 = 'X'
# filling list    
    while(game_on == True and j<=9):
        
        if j==9:
            print("Draw!")
            break
        # player 1
        game_list[position(game_list,choice_1,1)] = choice_1
        display(game_list)
        game_on = game_over(game_list,choice_1)
        j = j+1
        if j == 9:
            continue
        if game_on == False:
            print('player 1 won!')
            break
        #player 2
        game_list[position(game_list,choice_2,2)] = choice_2
        display(game_list)
        game_on = game_over(game_list,choice_2)
        j = j+1
        if j==9:
            continue
        if game_on == False:
            print('player 2 won!')
            break
    play = replay()
# poitions are based on typical nympad 
# 1 2 3
# 4 5 6
# 7 8 9
```
